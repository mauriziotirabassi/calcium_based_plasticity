```python
module_name, neuron_model_name, synapse_model_name = NESTCodeGeneratorUtils.generate_code_for(
    nestml_neuron_model= "waveform_neuron.nestml",
    nestml_synapse_model= "stdp_synapse.nestml",
    module_name= "nestml_wf_module",
    post_ports = ["post_spikes"],
    target_path = "target_wf")
```

# Now we can use the new module

```python
#os.environ['LD_LIBRARY_PATH'] = '/tmp/nestml_target:' + os.environ.get('LD_LIBRARY_PATH', '')
nest.ResetKernel()
nest.Install( "target_wf/nestml_wf_module")
nest.set_verbosity("M_ERROR")

delta_t = 15  # ms
duration = 10000 # ms
frequency = 1.0  # HzNESTCodeGeneratorUtils.generate_code_for()

neuron_model="waveform_neuron_nestml__with_stdp_synapse_nestml"
synapse_model="stdp_synapse_nestml__with_waveform_neuron_nestml"

pre_neuron = nest.Create(neuron_model, 1)
post_neuron = nest.Create(neuron_model, 1)

receptor_types = nest.GetDefaults(neuron_model)["receptor_types"]
print("Receptor types:", receptor_types, "\n")

recordables = nest.GetDefaults(neuron_model)['recordables']
print("recordables: \n", recordables,"\n")

print(nest.GetDefaults(neuron_model))
```

```python
wr = nest.Create("weight_recorder")
#nest.SetDefaults("static_synapse", {"weight_recorder": wr[0]})
nest.SetDefaults(synapse_model, {"weight_recorder": wr[0]})  # Attach recorder to the STDP synapse

syn_spec = nest.CollocatedSynapses({
                                   'synapse_model':synapse_model,
                                   'receptor_type': 2},
                                   {
                                   'synapse_model':synapse_model,     #w non influisce su recettori perchè w non è nella formula di g ampa ecc...
                                   'receptor_type': 3})

#nest.Connect(pre_neuron, post_neuron, "one_to_one", "static_synapse")
nest.Connect(pre_neuron, post_neuron, "one_to_one", syn_spec=syn_spec)
```

```python
conns = nest.GetConnections()
print(conns.get())

interval = 1000.0 / frequency
#spike_times = np.arange(interval, duration, interval)
spike_times=[1000]

if delta_t >= 0:
    delta_pre = 0.0
    delta_post = delta_t

else:
    delta_pre = -1 * delta_t
    delta_post = 0.0
    

spike_generator = nest.Create("spike_generator", {"spike_times": spike_times})

syn_spec_spike_pre = nest.CollocatedSynapses(
                                    {'synapse_model':'static_synapse',
                                     'receptor_type': 1,
                                     'weight': 999.0,
                                     'delay': 1.0 + delta_pre,
                                    })
syn_spec_spike_post = nest.CollocatedSynapses(
                                    {'synapse_model':'static_synapse',
                                     'receptor_type':1,
                                     'weight': 999.0,
                                     'delay': 1.0 + delta_post,
                                    })

nest.Connect(spike_generator, pre_neuron, "one_to_one", syn_spec=syn_spec_spike_pre)
nest.Connect(spike_generator, post_neuron, "one_to_one", syn_spec=syn_spec_spike_post)
```

```python
sd = nest.Create("spike_recorder")
nest.Connect(pre_neuron, sd)
nest.Connect(post_neuron, sd)

vm_pre = nest.Create("voltmeter")
vm_post = nest.Create("voltmeter")
nest.Connect(vm_pre, pre_neuron)
nest.Connect(vm_post, post_neuron)

#multimetro: p', 'cai_CR', 'cstar', 'I_Ca_vdcc', 'I_syn_nmda', 'I_syn_ampa',  'I_syn_exc'
mm = nest.Create('multimeter', 1, {'record_from': ['p','cai_CR','cai_CR_nmda','cai_CR_vdcc', 'cstar','I_Ca_vdcc', 'I_syn_nmda', 'I_syn_ampa',  'I_syn_exc', 'log_approx']})
nest.SetStatus(mm, {'interval': 0.1})
nest.Connect(mm, post_neuron)

#Simula
nest.Simulate(duration)
spikes = sd.get(['events'][0])

# Assuming mm, vm_pre, vm_post, spikes, and wr have already been set up correctly
fig, ax = plt.subplots(3, figsize=(18, 12))  # Increase to 4 subplots


ti= 950
tf = 2000

# Pre- and Post-synaptic potentials
ax[0].plot(vm_pre.events["times"], vm_pre.events["V_m"], color="b", label="Pre-synaptic potential")
ax[0].plot(vm_post.events["times"], vm_post.events["V_m"], color="orange", label="Post-synaptic potential")
ax[0].set_title("Pre- and Post-synaptic Membrane Potential")
ax[0].set_ylabel("V_m [mV]")
ax[0].set_xlim([ti, tf])
ax[0].legend()

# Raster plot
ax[1].set_title("Rasterplot")
ax[1].plot(spikes['times'][np.where(spikes['senders'] == 1)], spikes['senders'][np.where(spikes['senders'] == 1)], '.', markersize=40, alpha=0.5, color="b", label="Pre-synaptic spikes")
ax[1].plot(spikes['times'][np.where(spikes['senders'] == 2)], spikes['senders'][np.where(spikes['senders'] == 2)] - 1, '.', markersize=40, alpha=0.5, color="orange", label="Post-synaptic spikes")
ax[1].legend()
ax[1].set_ylabel("Neurons")
ax[1].set_yticklabels("")
ax[1].set_xlim([ti, tf])


# 'I_VDCC', 'I_AMPA', 'I_NMDA']} 'cstar', 'I_Ca_vdcc', 'I_syn_nmda', 'I_syn_ampa',  'I_syn_exc'
events = nest.GetStatus(mm, "events")[0]
times = events["times"]
I_Ca_vdcc = events["I_Ca_vdcc"]
I_Ca_nmda = events["I_syn_nmda"]
I_ampa = events["I_syn_ampa"]
ax[2].plot(times, I_Ca_vdcc, color="red")
ax[2].plot(times, I_Ca_nmda, color="blue")
ax[2].plot(times, I_ampa, color="green")
ax[2].set_title("Currents:VDCC red, NMDA blue, ampa green")
ax[2].set_ylabel("I [pA]")
ax[2].set_xlabel("Time [ms]")
ax[2].set_xlim([ti, tf])



fig.tight_layout()
plt.show()


fig, ax = plt.subplots(3, figsize=(18, 12))  # Increase to 4 subplots


# Currents
I_VDCC = events["I_Ca_vdcc"]
ax[0].plot(times, I_VDCC, color="r")
ax[0].set_title("I VDCC")
ax[0].set_ylabel("pA")
ax[0].set_xlabel("Time [ms]")
ax[0].set_xlim([ti, tf])

I_NMDA = events["I_syn_nmda"]
ax[1].plot(times, I_NMDA, color="b")
ax[1].set_title("I NMDA")
ax[1].set_ylabel("pA")
ax[1].set_xlabel("Time [ms]")
ax[1].set_xlim([ti, tf])


I_AMPA = events["I_syn_ampa"]
ax[2].plot(times, I_AMPA, color="g")
ax[2].set_title("I_AMPA ")
ax[2].set_ylabel("pA")
ax[2].set_xlabel("Time [ms]")
ax[2].set_xlim([ti, tf])

fig.tight_layout()
plt.show()

fig, ax = plt.subplots(3, figsize=(18, 12))  # Increase to 4 subplots

#Ca
cai_CR = events["cai_CR"]
cai_CR_vdcc = events["cai_CR_vdcc"]
cai_CR_nmda = events["cai_CR_nmda"]
ax[0].plot(times, cai_CR, color="g")
ax[0].plot(times, cai_CR_vdcc, color="r")
ax[0].plot(times, cai_CR_nmda, color="b")
ax[0].set_title("Calcium ")
ax[0].set_ylabel("Ca_in [mol/um3]")
ax[0].set_xlabel("Time [ms]")
ax[0].set_xlim([ti, tf])
print("max ca: ", np.max(cai_CR))
#print(cai_CR_vdcc[1200])

# C*
cstar = events["cstar"]
ax[1].plot(times, cstar, color="g")
ax[1].set_title("C*")
ax[1].set_ylabel("C* [mol/um3]ms")
ax[1].set_xlabel("Time [ms]")
ax[1].set_xlim([ti, tf])
print("mac c* :",np.max(cstar))

# p
ro = events["p"]
ax[2].plot(times, ro, color="r")
ax[2].set_title("ro")
ax[2].set_ylabel("ro")
ax[2].set_xlabel("Time [ms]")
ax[2].set_xlim([ti, tf])
```
